```python
#인공지능(대분류)
#
#-머신러닝: 기계학습, 데이터에서 규칙을 자동으로 학습(통계학 유래)
#ex) 사이킷런(scikit-learn)
#
#-딥러닝: 알고리즘 중 신경망을 기반으로 한 방법 통칭
#ex) 탠서플로우 
#
#-데이터분석
#입력데이터 분석의 기반, X로 표기(feature, 설명변수 등)
#출력데이터 모델을 통해 추정,예측하고자 하는 데이터,
#Y로 표시(종속변수, target, label, class등)
#
#--데이터 유형
#스칼라: 숫자
#벡터: 데이터레코드(ex. [1,0,1,0]), 괄호가 하나,1차원
#행렬: 벡터가 여러개인 데이터 집합(ex. [[1,2],[3,4]]),괄호가 2개,2차원
#텐서: 행렬이 여러개인 데이터 집합(ex. [[[1,2]]]), 괄호가 3개, 3차원
#
#--벡터의 연산
#같은 차원수의 벡터나 행렬은 덧셈과 뺄셈이 가능
#요소끼리 매핑되어 연산(element-wise 연산)
#ex) [1,2,3]+[2,2,2] = [3,4,5]
#
#백터에 스칼라 값을 곱하면 벡터의 모든 원소에 곱하는 것(나눗샘도 같음)
#ex)[1,2,3]*5 = [5,10,15]
#
#--백터의 내적 
#두 차원수가 같아야 함
#앞 백터 = 행, 뒤 백터 = 열
#같은 위치의 요소들끼리 곱한 후 총 합을 계산

#ex) [1,2,3] *  [5 = 1*5 + 2*6 + 3*3 = 26
#                6
#	               3]

#사이즈 결정
#(a,b) * (b,a) = (a,a) 
#> b가 같아야 곱하기 가능(공식)
#a는 결과값의 "크기"를 결정
```

```python
def cos_sim(a_v,b_v):
  res = np.dot(a_v, b_v) / (np.linalg.norm(a_v) * np.linalg.norm(b_v))
  return res #0.6826628516349054(ex)
```

```python
def get_gray_img(file_name):
  with open(f'/content/drive/MyDrive/Colab Notebooks/champion_image/{file_name}','rb') as f:
    res = f.read()
    get_gray_img = Image.open(BytesIO(res)).convert('L')
  return get_gray_img

def img2vec(img):
  img_vec = np.asarray(img, dtype=np.int).reshape(120*120)
  return img_vec
```

```python
dic = {}
for i in range(len(champion_file_lst)):
  img = get_gray_img(champion_file_lst[i])
  a = img2vec(img) 
  for j in range(i+1,len(champion_file_lst)):
    try:
      img2 = get_gray_img(champion_file_lst[j+1])
      b = img2vec(img2)
    except: continue
    cosres = cos_sim(a,b)
    dic[champion_file_lst[i].split('.')[0]+':'+champion_file_lst[j].split('.')[0]] = cosres
```

```python
#선생님 버전
rec_dic={}
for champ1 in champion_file_lst:
  for champ2 in champion_file_lst:
    if champ1 == champ2:
      continue
    
    #흑백이미지 가져오기
    champ1_img = get_gray_img(champ1)
    champ2_img = get_gray_img(champ2)

    #이미지를 벡터화
    champ1_vec = img2vec(champ1_img)
    champ2_vec = img2vec(champ2_img)

    #두 챔피언 코사인 유사도 구하기
    cos_res = cos_sim(champ1_vec, champ2_vec)
    rec_dic[champ1.split('.')[0]+':'+champ2.split('.')[0]] = cos_res
```
