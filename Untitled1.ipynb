```python
import pandas as pd
from tqdm import tqdm
import requests
```

```python
riot_api_key = 'RGAPI-82547be3-cc5e-4af4-9d63-d240fadc606d'
```

```python
def get_puuid(user):
    username_url = f'https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/{user}?api_key={riot_api_key}'
    userName = requests.get(username_url).json()
    userPuuid = userName['puuid']
    return userPuuid
```

```python
def get_matchid(userPuuid,num):
    usermatch_url = f'https://asia.api.riotgames.com/lol/match/v5/matches/by-puuid/{userPuuid}/ids?type=ranked&start=0&count={num}&api_key={riot_api_key}'
    userMatch= requests.get(usermatch_url).json()
    return userMatch
```

```python
def get_matches_timelines(userMatchid):
    usermatchdata_url = f'https://asia.api.riotgames.com/lol/match/v5/matches/{userMatchid}?api_key={riot_api_key}'
    matches = requests.get(usermatchdata_url).json()
    usertimeline_url = f'https://asia.api.riotgames.com/lol/match/v5/matches/{userMatchid}/timeline?api_key={riot_api_key}'
    timeline = requests.get(usertimeline_url).json()
    return matches, timeline
```
