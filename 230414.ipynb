```python
import pandas as pd 
import numpy as np 
```

```python
#데이터 전처리 
#KNN모델 사용  확률계산
#확률 확인
```

```python
fish = pd.read_csv('https://bit.ly/fish_csv_data')
```

```python
fish.head()
```

```python
pd.unique(fish.Species) #컬럼의 값이 몇 종류 있는지 확인 가능 
```

```python
fish_input = fish[['Weight','Length','Diagonal','Height','Width']]
fish_input.head()
```

```python
fish_input = fish_input.to_numpy()
```

```python
fish_input[:5]
```

```python
fish_target = fish.Species.to_numpy()
fish_target[:5]
```

```python
len(fish)
```

```python
#데이터 나누기
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target= train_test_split(fish_input, fish_target, test_size =0.2 , random_state = 42)
#train input 80%값 
#test input 80% 값
#train_target 80%의 결과
#test_target 20%의 결과 
#test size 데이터 값중 0.2 = 20%만 사용 함
```

```python
train_input.shape
```

```python
test_input.shape #두개의 합 159 = len(fish)
#5는 특성, 차원 수 
```

```python
#표준 점수 정규화
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler() #표준점수 정규화
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
train_input[:5]
#데이터 차이가 큰 폭은 모델이 방향을 못잡음(성능저하)
```

```python
train_scaled[:5]
#표준점수를 반영해서 정규과정을 거침
```

```python
#모델 선언 KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
train_target
```

```python
kn = KNeighborsClassifier(n_neighbors = 3)
kn.fit(train_scaled, train_target)
```

```python
kn.score(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

```python
kn.score(train_scaled, train_target)
```

```python
kn.score(test_scaled, test_target)
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
bream_smelt_indexes = (train_target =='Bream') | (train_target == 'Smelt') #true of false로 해당 값 가져오기 
bream_smelt_indexes
```

```python
train_bream_smelt = train_scaled[bream_smelt_indexes]
target_bream_smelt = train_target[bream_smelt_indexes]
```

```python
target_bream_smelt
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(train_bream_smelt, target_bream_smelt)
```

```python
lr.classes_
```

```python
from scipy.special import expit 
#시그노이드 함수
```

```python
decisions = lr.decision_function(train_bream_smelt[:5])
#로지스틱스는 음수도 가지게 됨 
```

```python
decisions #중간값, 실제있는 값들로 시그모이드 넣을 것 
```

```python
expit(decisions) 
#시그모이드 특징 0에서 1의 값만 가진다. 
#비선형 결과 값 
```

```python
#다중 분류 문제 해결
#파라미터값 C(정규화정도(데이터 특성에 따라 범위지정해서 정규화 진행), 하이퍼파라미터), max_iter
lr = LogisticRegression(C=20,max_iter=1000)
```

```python
lr.fit(train_scaled, train_target)
```

```python
lr.score(train_scaled, train_target)
```

```python
lr.score(test_scaled, test_target)
```

```python
lr = LogisticRegression(C=2,max_iter=1000)
lr.fit(train_scaled, train_target)
lr.score(train_scaled, train_target), lr.score(test_scaled, test_target)
```

```python
lr = LogisticRegression(C=50,max_iter=1000)
lr.fit(train_scaled, train_target)
lr.score(train_scaled, train_target), lr.score(test_scaled, test_target)
```

```python
lr = LogisticRegression(C=20,max_iter=10)
lr.fit(train_scaled, train_target)
lr.score(train_scaled, train_target), lr.score(test_scaled, test_target)
```

```python
lr = LogisticRegression(C=2,max_iter=100)
lr.fit(train_scaled, train_target)
lr.score(train_scaled, train_target), lr.score(test_scaled, test_target)
```

```python
dec = lr.decision_function(test_scaled[:5])
```

```python
dec
```

```python
#소수점 제거
dec = np.round(dec,2)
dec
```

```python
#소프트맥스: 모든 값의 총합을 1이 되게 출력
from scipy.special import softmax
```

```python
proba = softmax(dec, axis=1)
```

```python
proba = np.round(proba,2)
```
